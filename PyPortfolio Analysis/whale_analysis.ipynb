# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

* `whale_returns.csv`: Contains returns of some famous "whale" investors' portfolios.

* `algo_returns.csv`: Contains returns from the in-house trading algorithms from Harold's company.

* `sp_tsx_history.csv`: Contains historical closing prices of the S&P TSX 60 Index.

## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale returns, convert the dates to a DateTimeIndex
whalepath=Path("../Resources/whale_returns.csv")
whale_df = pd.read_csv(whalepath,  index_col="Date", infer_datetime_format= True, parse_dates = True)
whale_df.head()
```

```python
# Count nulls
whale_df.info()
# Check % of null in each column
whale_df.isnull().mean()*100
# Check duplicated data 
whale_df.duplicated()
```

```python
# Drop nulls
whale_df=whale_df.dropna().copy()
whale_df
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
algopath=Path("../Resources/algo_returns.csv")
algo_df = pd.read_csv(algopath, index_col="Date", infer_datetime_format= True, parse_dates = True)
algo_df.head()
```

```python
# Count nulls
algo_df.info()
# Check % of null in each column
algo_df.isnull().mean()*100
# Check duplicate data 
algo_df.duplicated()
```

```python
# Drop nulls
algo_df=algo_df.dropna().copy()
algo_df.head()
```

## S&P TSX 60 Returns

Read the S&P TSX 60 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P TSX 60 Closing Prices
tsxpath = Path("../Resources/sp_tsx_history.csv")
tsx_df = pd.read_csv(tsxpath)
# Set the date as the index to the DataFrame
tsx_df.set_index(pd.to_datetime(tsx_df['Date'], infer_datetime_format=True), inplace=True)
tsx_df.drop(columns=["Date"], inplace = True)
tsx_df.head(1)
```

```python
# Check % of null in each column
tsx_df.isnull().mean()*100
```

```python
# Check Data Types
tsx_df.dtypes
```

```python
# Fix Data Types of "$ and ."
tsx_df["Close"]=tsx_df["Close"].str.replace("$", "").copy()
tsx_df["Close"]=tsx_df["Close"].str.replace(",", "").copy()
tsx_df.head()
```

```python
# Turns object to float first
tsx_df["Close"] = tsx_df["Close"].astype('float')
# Calculate Daily Returns
tsx_return_df=tsx_df.pct_change()
tsx_return_df.head()
```

```python
# Drop nulls
tsx_return_df= tsx_return_df.dropna().copy()
tsx_return_df.head()
```

```python
# Rename `Close` Column to be specific to this portfolio.
tsx_return_df.columns = ["S&P TSX 60"]
tsx_return_df.head()
```

## Combine Whale, Algorithmic, and S&P TSX 60 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P TSX 60 Returns into a single DataFrame with columns for each portfolio's returns.

join_col_df  = pd.concat([whale_df, algo_df, tsx_return_df], axis="columns", join="inner")
join_col_df.head()
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Calculate and plot the rolling standard deviation for the S&P TSX 60 using a 21-day window
2. Calculate the correlation between each stock to determine which portfolios may mimick the S&P TSX 60
3. Calculate and plot a 60-day Beta for Berkshire Hathaway Inc compared to the S&P 60 TSX

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
rolling_std = join_col_df.rolling(window = 21).std().copy()
# Plot the rolling standard deviation
rolling_std.plot(figsize=(20,10))
```

### Calculate and plot the correlation

```python
# Calculate the correlation
portfolio_corr = join_col_df.corr()
# Display de correlation matrix
portfolio_corr
```

```python
#Display heatmap of correlation
sns.heatmap(portfolio_corr, vmax = 1, vmin = -1)
```

### Calculate and Plot Beta for a chosen portfolio (SOROS FUND MANAGEMENT LLC) and the S&P 60 TSX

```python
# Calculate covariance of a single portfolio - SOROS FUND MANAGEMENT LLC
soros_cov= join_col_df["SOROS FUND MANAGEMENT LLC"].cov(join_col_df["S&P TSX 60"])
soros_cov

# Calculate variance of S&P TSX
tsx_return_var=join_col_df["S&P TSX 60"].var()
tsx_return_var
# Computing beta = cov(s, m)/ var(m)
soros_beta = soros_cov / tsx_return_var
print (f' Beta of SOROS FUND MANAGEMENT LLC is {soros_beta}')
```

```python
# Calculate 21-day rolling beta of SOROS FUND MANAGEMENT LLC and plot the data
#rolling cov of SOROS FUND MANAGEMENT LLC
rolling_soro_covariance=join_col_df["SOROS FUND MANAGEMENT LLC"].rolling(window = 21).cov(join_col_df["S&P TSX 60"])
rolling_tsx_return_var = join_col_df["S&P TSX 60"].rolling(window=21).var()

rolling_soro_beta = rolling_soro_covariance / rolling_tsx_return_var
rolling_soro_beta.plot(figsize=(20, 10), title='Rolling 21-Day Beta of SOROS FUND MANAGEMENT LLC')
```

## Rolling Statistics Challenge: Exponentially Weighted Average 

An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the [`ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html) with a 21-day half-life.

```python
# Use `ewm` to calculate the rolling window
portfolio_exponentially_weighted_average=join_col_df.ewm(halflife=21).mean().plot(figsize=(20, 10),
                                                                                  title='Rolling 21-Day EWA of all portfolio' )
```
